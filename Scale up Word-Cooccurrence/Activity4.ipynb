```r
library(ggplot2)
```

```r
y<-c(0.0563,0.15590,0.09552,0.30692,2.5714,0.33702,0.58746,0.39854,3.7526)
x<-c(5,10,15,20,25,30,35,40,45)
```

```r
data<-data.frame(x,y)
```

```r
head(data)
```

```r
qplot(x, y, data=data,xlab='file',ylab='time') 
```

```r
x1<-c(5,10,15,20,25,30,35,40,45,50)
y1<-c(0.37982,4.2595,1.48239,3.1423,2.5507,1.5757,6.6376,5.2583,6.169,7.26)
data2<-data.frame(x1,y1)
```

```r
qplot(x1, y1, data=data2,xlab='file',ylab='time') 
```

```r
ggplot(data, aes(x,y)) + geom_line(colour = 'blue')+ geom_point(colour='blue')+ xlab("Files") +
  ylab("Time") + ggtitle("For neighbours = 2") + geom_text(aes(label=y),hjust=0, vjust=0)+ scale_x_continuous(limits = c(0, 70),breaks=c(0,10,20,30,40,50)) + scale_y_continuous(limits = c(0,5))
```

```r
ggplot(data2, aes(x1,y1)) + geom_line(colour = 'blue')+ geom_point(colour='blue')+ xlab("Files") +
  ylab("Time") + ggtitle("For neighbours = 3") + geom_text(aes(label=y1),hjust=0, vjust=0)+ scale_x_continuous(limits = c(0, 70),breaks=c(0,10,20,30,40,50)) + scale_y_continuous(limits = c(0,9))
```
